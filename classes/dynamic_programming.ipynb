## 1. definicion:

Programacion dinamica es un paradigma de resolucion de problemas.  
Esta basado en el almacenamiento de sub-problemas similares al  
problema en si.  
    
Hay 2 grandes clases, los problemas combinatorios (cantidad de soluciones)  
y los problemas de optimizacion (maximizar o minimizar una solucion).

## 2. Caracterizacion:

###    1.- solucion parcial debe ser posible:

        - Se debe poder armar una solucion total
        a partir de una construccion parcial del 
        problema.

        ejemplos: 
            1. armar un rompecabezas! Poner una pieza 
            te acerca cada vez mas a la solucion, 
            podriamos ver este problema como que
            queremos ver si existe una configuracion 
            posible de las piezas, y la respesta seria
            si se puede encontrar una configuracion
            con todas las piezas.

            2. el problema de la mochila. si una 
            mochila tiene capacidad M, y agrego
            un objeto de volummen c, entonces,
            el problema es el mismo pero con
            capacidad M-c (el problema se repite 
            pero para un caso parcial).

        esto nos dice que para hallar la respuesta
        para un estado del problema, necesitamos
        procesar estados mas peque単os para dar
        respuesta a este. debe ser una solucion recursiva!
        
###    2.- sub-estructura optima:

        - La sub-estructura optima, que se necesita para
        problemas de optimizacion, se debe cumplir
        para que sea feasible una solucion con 
        programacion dinamica.
    
        ejm:
            1. (sub-estructura optima) hallar la minima 
            cantidad de monedas para dar vuelto igual a m.

            2. (no sub-estructura optima) hallar el 
            maximo & de un rango en un array.
        
        nota: si al modelar un problema no encuentro una
            sub-estructura optima no necesariamente quiere
            decir que no exista otra caracterizacion con
            sub-estructura optima. 

###    3.- Memoization y sobrelapamiento de subproblemas:

        - Como se necesita guardar informacion, entonces
        el proceso en el cual calculamos un estado y guardamos
        ese resultado para luego poderse usar se llama 
        memoization. Tenga en cuenta que de nada serviria 
        guardar los resultados de los estados del problema 
        si estos no se usan despues! Por eso, un problema
        de programacion dinamica incluye tambien un problema
        de sobrelapamiento!

        ejemplo:
            fibonacci: calcular f(n) y f(n-1) tienen f(n-2)
            en comun! "No haga doble trabajo, guarde!".

        nota: a pesar de esto hay problemas en los cuales
        es imposible guardar los estados (por la cantidad), 
        en este caso una solucion greedy o de teoria de juegos 
        podria servir, quizas fuerza bruta.

## 3. tipos de implementacion:

	Hay dos tipos de implementacion de un algoritmo
	de programacion dinamica, Bottom up y top down.

	Botttom up:
	
	Se refiere a la construccion natural de los estados
	en orden a complejidad. Calcular fibonacci en orden	
	creciente de n, es natural, ya que necesitas de un 
	n menor para calcular un n mayor.

	pro: son mas eficientes y se pueden hacer algunas mejoras
	de memoria significativas.

	contra: en ocasiones generas mas estados de los que
	necesitaras, ya que generas todos los estados menos
	complejos hasta llegar al tuyo.

	nota: para el caso de fibonacci obviamente necesito
	hallar todos los n menores para calcular el actual,
	pero si la recursion fuera por ejemplo: f(n) = 
	f(n - 10) + f(n - 9) es muy probable que no me 
	sirvan los estados f(n-8), ..., f(n-1).

	Top-down:

	Esta guiado por una solucion backtracking, en general
	son mas faciles de codear y de entender, ya que se
	construyen segun lo que vayas necesitando. 

	pro: es bien intuitivo de codear, pasar de una solucion
	de backtracking a una de dp es muy simple, recorres
	menos estados ya que solo usas los justos.

	contra: como es recursivo, usa mas recursos para
	llamar a cada funcion, y suelen gastar mas memoria
	siendo a veces muy poco practico haciendo optimizaciones
	de memoria.

## 4. ejemplos:

    * fibonacci:
    
        - halle fibonacci(n) (mod 1000000007)
    
    solucion:
    
    Conocemos gracias a los problemas de recursion, que este problema
    puede solucionarse con el siguiente codigo, pero para casos 
    peque単os.
    
```cpp
const int mod = 1e9 + 7;
int fibonacci(int n) {
    if (n <= 1) return n;
    return fibonacci(n-1) + fibonacci(n-2);
}
```
    
    p: Puedo usar programacion dinamica para resolver este problema?
    
    r: claro que si, primero veamos que este problema tiene una recurrencia
    natural (quiere decir que se puede armar parcialemente), por otro lado
    ya habiamos afirmado que este problema tiene estados sobrelapados, es 
    facil ver que estos estados son O(n) uno para cada fibonacci(i).
    
    como se veria el anterior codigo con dp?

```cpp
const int mod = 1e9 + 7;
const int maxN = 1e6 + 10;
bool vis[maxN];
int memo[maxN];
int fibonacci(int n) {
    if (n <= 1) return n;
    if (vis[n]) return memo[n];
    vis[n] = 1;
    return memo[n] = (fibonacci(n-1) + fibonacci(n-2)) % mod;
}
```

    * knapsack problem 0/1

        - dado n objetocs con peso wi y valor vi, y una mochila con capacidad C,
        hallar que objetos debo escojer para maximizar el valor.

    * min-coin: (knapsack problem 0/infinity simplificado)

        - dado n tipos de monedas, y un valor m, cual es la minima cantidad de monedas
        necesarias tal que tengan suma de valores igual a m.

    * win-lose

        - Recibes una torre de n bloques por tu cumplea単os, y quieres jugar con un  
        amigo el siguiente juego: Juegas por turnos, y en cualquier turno cada persona 
        puede quitar 1, 3, 7 bloques de la cima, si gana el que se lleva el bloque de 
        mas abajo, puedes decir quien ganara si los dos juegan optimamente y tu amigo 
        empieza?

## 5. conclusiones:

    - los problemas de programacion dinamica se pueden resolver con backtracking, pero
    para que se pueda resolver un problema de backtracking con programacion dinamica,
    deben: sobrelaparse estados, tener restricciones "peque単as" (para poder guardar estados)
    y si es un problema de optimizacion, se debe cumplir la sub-estructura optima.
